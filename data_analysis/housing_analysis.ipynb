The following is analysis of US Census housing data.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../data/homes_by_state.csv')
```

```python
df.head()
```

```python
df['Percent No Vehicles'].hist()
```

```python
vehicle_sort = df.sort_values('Percent No Vehicles')
```

```python
vehicle_sort
```

With the notable exception of Alaska, there seem to be two main reasons for states to have low car ownerships: either they are highly urban or dominated by a large city, or they are states with lower average incomes. Geographically, this translates to many states in the northeast appearing high on the list,as well as several southern and midwestern states; the states with higher vehicle ownership tend to be those that are more rural.

```python
df['Median Home Value'].hist()
```

```python
value_sort = df.sort_values('Median Home Value')
```

```python
value_sort
```

Like many of the other indicators I've looked at, there is a strong lean towards the northeast and the west coast with the median home values. This is another indication of the desirability of an area - although in the case of Hawaii, for example, that desirability is hampered by the sheer lack of housing stock, raising the home values.

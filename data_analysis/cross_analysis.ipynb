The following is analysis of all four other metrics.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
commute_df = pd.read_csv('../data/commute_mode_by_state.csv')
homes_df = pd.read_csv('../data/homes_by_state.csv')
income_df = pd.read_csv('../data/income_data_by_state.csv')
move_df = pd.read_csv('../data/move_by_state.csv')
```

```python
homes_df['Rate Public Transit'] = commute_df['Rate Public Transit']
homes_df['Median Income'] = income_df['Median Income']
homes_df['Moved Total'] = move_df['Moved Total']
```

```python
homes_df.head()
```

This is the combined table of the metrics we want.

```python
sns.lmplot(x='Rate Public Transit', y='Percent No Vehicles', data = homes_df)
plt.title('Public Transit Usage vs. Percent No Vehicles')
```

A graph of public transit usage versus the percentage of people who don't own vehicles in each state. As expected, there is a positive correlation here.

```python
sns.lmplot(x='Rate Public Transit', y='Median Home Value', data = homes_df)
plt.title('Public Transit Usage vs. Median Home Value')
```

A graph of public transit usage versus median home values in each state. There is a positive correlation here, suggesting that the presence of transit increases home values.

```python
sns.lmplot(x='Rate Public Transit', y='Median Income', data = homes_df)
plt.title('Public Transit Usage vs. Median Income')
```

A graph of public transit usage versus median income in each state. There is a positive correlation here, suggesting that the presence of transit increases income.

```python
sns.lmplot(x='Rate Public Transit', y='Moved Total', data = homes_df)
plt.title('Public Transit Usage vs. Out-of-State Move Ins')
```

A graph of public transit usage versus out-of-state move ins in each state. There is practically no correlation here, suggesting that the presence of transit does NOT impact people moving in from other parts of the country.
